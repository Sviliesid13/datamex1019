# Supervised classification

In the data.csv there are letters (uppercases and lowercases) and numbers, 28x28 pixels in a row format.

* First, you need to know which labels are which, meaning you need to visualize some data to realize which number labels represents a letter, or a number.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Your code here
df = pd.read_csv('data_all.zip')
```

```python
matriz = df.values
for i in range(len(df['e'].unique())):
    plt.imshow(matriz[i, 1:].reshape(28,28))
    plt.show()
    print(df['e'][i])
```

* Now, try to train a classifier model to predict the uppercases. Use every single model you know for classification.

```python
# Your code here
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge

uppercases = df[(df['e'] < 36) & (df['e'] > 9)]
X = uppercases.iloc[:, 1:]
y = uppercases.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
# LinearRegression

linreg = LinearRegression().fit(X_train, y_train)
linreg.score(X_train, y_train)
```

```python
# Ridge

ridge = Ridge().fit(X_train, y_train)
ridge.score(X_train, y_train)
```

* Try to do the same thing with lowercases.

```python
# Your code here
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import Lasso

lowercases = df[df['e'] > 36]
X = lowercases.iloc[:, 1:]
y = lowercases.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
# RidgeCV

ridgecv = RidgeCV().fit(X_train, y_train)
ridgecv.score(X_train, y_train)
```

```python
# Lasso

lasso = Lasso().fit(X_train, y_train)
lasso.score(X_train, y_train)
```
